```python
import cv2
from matplotlib import pyplot as plt
```

```python
def showCV(image,title):
    cv2.imshow(title, image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

def showplt(image):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image)
```

## Countours
contours are continious lines or curves that bound or cover the full boundary of image


```python
img = cv2.imread('coin.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
temp,img1 =  cv2.threshold(img,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
coutours, hierarhy = cv2.findContours(img1,cv2.RETR_LIST,cv2.CHAIN_APPROX_NONE)
cv2.drawContours(img1,coutours,-1,(0,255,0),thickness=1)
showplt(img1)
```

### line circle and blob detection 
The hough trasnform takes binary edge map as input and attempts to locate edges placed as straight lines. the idea of the hough trasnform is that every dge point in the edge map is tranformed to all possible lines that could pass thorgh the point

```python
from vision_utils import showCV,showPlt
import cv2
import numpy as np

img = cv2.imread("leena.jpg",0)
edges = cv2.Canny(img,100,170,apertureSize=3)

lines = cv2.HoughLines(edges,1,np.pi/180,240)
print(lines)
```

```python
## circle detection 
import cv2
imgnew = cv2.imread("testcircles.jpg")
imgnew = cv2.cvtColor(imgnew, cv2.COLOR_RGB2GRAY)
blur = cv2.medianBlur(imgnew,5)

circles = cv2.HoughCircles(blur,cv2.HOUGH_GRADIENT,1.2,30)

```

```python
showPlt(imgnew)
```

```python
circles = np.uint16(np.around(circles))
for x,y,r in circles[0,:]:
    cv2.circle(imgnew, (x, y), r, (0, 255, 0), 2)  # Circle outline
    cv2.circle(imgnew, (x, y), 2, (0, 0, 255), 3)  # Center point
```

```python
showPlt(imgnew)
```

## Finging corner


```python
img = cv2.imread("chessboard.jpg")

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = np.float32(gray)

harris_corners = cv2.cornerHarris(gray,2,3,0.04)

# dialate the corners 
kernel = np.ones((7,7),dtype="uint8")
harris_corners = cv2.dilate(harris_corners,kernel=kernel,iterations=2)
img[harris_corners>0.01*harris_corners.max()] = [255,127,127]
showPlt(img)


```
