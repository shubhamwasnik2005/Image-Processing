## yolov3 in opencv

```python
import cv2
labelsPath = "coco.names"
Lables = open(labelsPath).read().split("\n")
```

```python
weightPth = "yolov3.weights"
cfg_path = "yolov3.cfg"


net = cv2.dnn.readNetFromDarknet(cfg_path,weightPth)
net.setPreferableBackend(cv2.dnn.DNN_BACKEND_OPENCV)

ln = net.getLayerNames()
print(ln)
```

```python
imagepth = "lambosample.jpg"
img = cv2.imread(imagepth)
blob = cv2.dnn.blobFromImage(image=img,scalefactor=1/255.0,size=(416,416),swapRB=True,crop=False)

```

```python
net.setInput(blob=blob)
layeroutput=net.forward(ln)
```

```python
for output in layeroutput:
    print(output)
```
