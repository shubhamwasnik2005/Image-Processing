```python
import cv2
from matplotlib import pyplot as plt
```

```python
def showCV(image,title):
    cv2.imshow(title, image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

def showplt(image):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image)
```

### Histogram
1) View histogram representation of images
2) use k means clustering to get the dominant colors and their proportions in images

```python
image = cv2.imread("lambosample.jpg")
plt.hist(image.ravel(),256,[0,255])
plt.show()
```

```python
#showing histogram of seperate color channel
color = ('b','g','r')
for i, col in enumerate(color):
    histogram2= cv2.calcHist([image],[i],None,[256],[0,256])
    plt.plot(histogram2,color=col)
```

## k means clustering to find to dominant colors in the images

```python
from sklearn.cluster import KMeans
imgpth = "lambosample.jpg"
image =cv2.imread(imgpth)
image = image.reshape((image.shape[0]*image.shape[1]),3)

clusters = 5
clt = KMeans(clusters)
clt.fit(image)
clt
```

## Comparing images with MSE

```python
import numpy as np
def mse(image1,image2):
    error = np.sum((image1.astype("float")-image2.astype("float"))**2)
    error /= float(image1.shape[0]*image2.shape[1])
    return error
```

```python
image1 = cv2.imread("lambosample.jpg")
image2 = cv2.GaussianBlur(image1,(7,7),0)
mse(image1,image2)
```

### Filtering colors


```python
lower = np.array([0,0,0])
upper = np.array([10,255,255])

hsv_img = cv2.cvtColor(image1,cv2.COLOR_BGR2HSV)
mask = cv2.inRange(hsv_img,lower,upper)


res = cv2.bitwise_and(image1,image1,mask=mask)
showplt(res)
```
