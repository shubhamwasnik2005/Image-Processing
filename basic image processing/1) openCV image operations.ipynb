## Showing image using cv2

```python
def image_showing(pth):
    img = cv2.imread(pth)
    cv2.imshow("test", img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
import cv2
from matplotlib import pyplot as plt

def showCV(img,nme):
    nme = "".format(nme)
    cv2.imshow(nme,img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

def showplt(img):
    img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    plt.imshow(img)
```

## Transformation of images
1. perform image translation
2. rotation
3. rotation with transpose
4. flipping image

```python
imgpth = "leena.jpg"
img = cv2.imread(imgpth)
ht,wd = img.shape[:2]
quarter_ht,quarter_wd = ht/4,wd/4

T = np.array([[1,0,quarter_ht],[0,1,quarter_ht]])
img_trasform = cv2.warpAffine(img,T,(wd,ht))
showplt(img_trasform)
```

```python
## Get Rotation
rotation_mtx = cv2.getRotationMatrix2D((wd/2,ht/4),90,1)
cv2.getRotationMatrix2D()
rotimg = cv2.warpAffine(img,rotation_mtx,(wd,ht))

flip = cv2.flip(img,1)
showplt(flip)
```

## Bit wise operation

```python
square = np.zeros((300,300),dtype="uint8")
cv2.rectangle(square,(50,50),(250,250),255,-2)
plt.imshow(square)

ellipse = np.zeros((300,300),dtype="uint8")
cv2.ellipse(ellipse,(150,150),(150,150),30,0,180,255,-1)
plt.imshow(ellipse)
```

```python
andBit = cv2.bitwise_and(square,ellipse)
cv2.imshow("and",andBit)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Dilation, Erosion and edge detection


```python
import cv2
import numpy as np
imgPath = "lambosample.jpg"
image = cv2.imread(imgPath,0)
kernel = np.ones((5,5),dtype = "uint8")

# erode 
erosion = cv2.erode(image,kernel=kernel)

# dilate
dialate = cv2.dilate(image,kernel=kernel)
showplt(dialate)

#opening - good for removing noise form the image erossion followed by dialtion
opening = cv2.morphologyEx(image,cv2.MORPH_OPEN,kernel=kernel)

#closing - goor for removing noise dialtion followed by erossion
closing = cv2.morphologyEx(image,cv2.MORPH_CLOSE,kernel)

showplt(closing)
```
